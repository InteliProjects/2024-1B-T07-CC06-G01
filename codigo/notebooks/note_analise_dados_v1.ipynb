```python
# copy data
data_copy = data.copy()

# translate the latitude and longitude so that the minimum latitude and longitude are 0
data_copy['LATITUDE'] = data_copy['LATITUDE'] - data_copy['LATITUDE'].min()
data_copy['LONGITUDE'] = data_copy['LONGITUDE'] - data_copy['LONGITUDE'].min()

# plot the translated latitude and longitude
plt.scatter(data_copy['LATITUDE'], data_copy['LONGITUDE'], alpha=0.5)
plt.title('Plotagem de Latitude vs Longitude')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

```

```python
def degrees_to_radians(degrees):
    return degrees * np.pi / 180

def convert_to_cartesian(latitude, longitude):
    lat_rad = degrees_to_radians(latitude)
    lon_rad = degrees_to_radians(longitude)
    x = np.cos(lat_rad) * np.cos(lon_rad)
    y = np.cos(lat_rad) * np.sin(lon_rad)
    return x, y

def euclidian_distance(x1, y1, x2, y2):
    return np.sqrt((x1 - x2)**2 + (y1 - y2)**2)
```

```python
# create x and y columns
data_copy['LAT_RAD'] = data_copy['LATITUDE'].apply(degrees_to_radians)
data_copy['LON_RAD'] = data_copy['LONGITUDE'].apply(degrees_to_radians)

# create x and y columns
data_copy['X'], data_copy['Y'] = zip(*data_copy.apply(lambda row: convert_to_cartesian(row['LAT_RAD'], row['LON_RAD']), axis=1))

# plot the translated latitude and longitude
plt.scatter(data_copy['X'], data_copy['Y'], alpha=0.5)
plt.title('Plotagem de X vs Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

# Calculate distance matrix
dist_matrix = euclidean_distances(data_copy[['X', 'Y']])

def nearest_neighbor(dist_matrix):
    n = len(dist_matrix)
    visited = [False] * n
    tour = [0]  # starting from node 0

    while len(tour) < n:
        current_node = tour[-1]
        min_dist = float('inf')
        nearest_node = None

        for neighbor in range(n):
            if not visited[neighbor] and dist_matrix[current_node][neighbor] < min_dist:
                min_dist = dist_matrix[current_node][neighbor]
                nearest_node = neighbor

        tour.append(nearest_node)
        visited[nearest_node] = True

    # Add back the starting node to complete the tour
    tour.append(tour[0])
    
    return tour

tsp_tour = nearest_neighbor(dist_matrix)
```

```python
from ortools.constraint_solver import routing_enums_pb2
from ortools.constraint_solver import pywrapcp
```

```python
# [START data_model]
def create_data_model():
    """Stores the data for the problem."""
    data = {}
    # Locations in block units
    data["locations"] = locations
    print("Numero de pontos:", len(data["locations"]))
    data["num_vehicles"] = 1
    data["depot"] = 0
    return data
    # [END data_model]

# [START distance_callback]
def compute_euclidean_distance_matrix(locations):
    """Creates callback to return distance between points."""
    distances = {}
    for from_counter, from_node in enumerate(locations):
        distances[from_counter] = {}
        for to_counter, to_node in enumerate(locations):
            if from_counter == to_counter:
                distances[from_counter][to_counter] = 0
            else:
                # Euclidean distance
                distances[from_counter][to_counter] = int(
                    math.hypot((from_node[0] - to_node[0]), (from_node[1] - to_node[1]))
                )
    return distances
    # [END distance_callback]

# [START solution_printer]
def print_solution(manager, routing, solution):
    """Prints solution on console."""
    print(f"Funcao Objetivo: {solution.ObjectiveValue()}")
    index = routing.Start(0)
    plan_output = "Rota:\n"
    route_distance = 0
    while not routing.IsEnd(index):
        plan_output += f" {manager.IndexToNode(index)} ->"
        previous_index = index
        index = solution.Value(routing.NextVar(index))
        route_distance += routing.GetArcCostForVehicle(previous_index, index, 0)
    plan_output += f" {manager.IndexToNode(index)}\n"
    print(plan_output)
    plan_output += f"Distancia percorrida: {route_distance}m\n"
    # [END solution_printer]

def main():
    """Entry point of the program."""
    # Instantiate the data problem.
    # [START data]
    data = create_data_model()
    # [END data]

    # Create the routing index manager.
    # [START index_manager]
    manager = pywrapcp.RoutingIndexManager(
        len(data["locations"]), data["num_vehicles"], data["depot"]
    )
    # [END index_manager]

    # Create Routing Model.
    # [START routing_model]
    routing = pywrapcp.RoutingModel(manager)
    # [END routing_model]

    # [START transit_callback]
    distance_matrix = compute_euclidean_distance_matrix(data["locations"])

    def distance_callback(from_index, to_index):
        """Returns the distance between the two nodes."""
        # Convert from routing variable Index to distance matrix NodeIndex.
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return distance_matrix[from_node][to_node]

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)
    # [END transit_callback]

    # Define cost of each arc.
    # [START arc_cost]
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)
    # [END arc_cost]

    # Setting first solution heuristic.
    # [START parameters]
    search_parameters = pywrapcp.DefaultRoutingSearchParameters()
    search_parameters.first_solution_strategy = (
        routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC
    )
    # [END parameters]

    # Solve the problem.
    # [START solve]
    solution = routing.SolveWithParameters(search_parameters)
    # [END solve]

    # Print solution on console.
    # [START print_solution]
    if solution:
        print_solution(manager, routing, solution)
    # [END print_solution]
```

```python
main()
```

```python
from ortools.constraint_solver import routing_enums_pb2
from ortools.constraint_solver import pywrapcp

def create_data_model(dist_matrix, num_vehicles, depot):
    data = {}
    data['distance_matrix'] = dist_matrix.tolist()
    data['num_vehicles'] = num_vehicles
    data['depot'] = depot
    return data

def print_solution(data, manager, routing, solution):
    """Prints solution on console."""
    print(f"Objective: {solution.ObjectiveValue()}")
    max_route_distance = 0
    for vehicle_id in range(data["num_vehicles"]):
        index = routing.Start(vehicle_id)
        plan_output = f"Route for vehicle {vehicle_id}:\n"
        route_distance = 0  # Initialize route distance inside the loop
        while not routing.IsEnd(index):
            plan_output += f" {manager.IndexToNode(index)} -> "
            previous_index = index
            index = solution.Value(routing.NextVar(index))
            route_distance += routing.GetArcCostForVehicle(
                previous_index, index, vehicle_id
            )  # Update route distance correctly
        plan_output += f"{manager.IndexToNode(index)}\n"
        plan_output += f"Distance of the route: {route_distance}m\n"
        print(plan_output)
        max_route_distance = max(route_distance, max_route_distance)
    print(f"Maximum of the route distances: {max_route_distance}m")


def routing_problem(data):
    def distance_callback(from_index, to_index):
        from_node = manager.IndexToNode(from_index)
        to_node = manager.IndexToNode(to_index)
        return data['distance_matrix'][from_node][to_node]

    manager = pywrapcp.RoutingIndexManager(len(data['distance_matrix']), data['num_vehicles'], data['depot'])
    routing = pywrapcp.RoutingModel(manager)

    transit_callback_index = routing.RegisterTransitCallback(distance_callback)

    # Define cost of each arc.
    routing.SetArcCostEvaluatorOfAllVehicles(transit_callback_index)

    # Add Distance constraint.
    dimension_name = "Distance"
    routing.AddDimension(
        transit_callback_index,
        0,  # no slack
        10000,  # vehicle maximum travel distance
        True,  # start cumul to zero
        dimension_name,
    )
    distance_dimension = routing.GetDimensionOrDie(dimension_name)
    distance_dimension.SetGlobalSpanCostCoefficient(100)

    search_parameters = pywrapcp.DefaultRoutingSearchParameters()
    search_parameters.first_solution_strategy = (routing_enums_pb2.FirstSolutionStrategy.PATH_CHEAPEST_ARC)

    solution = routing.SolveWithParameters(search_parameters)

    # Print solution on console.
    if solution:
        print_solution(data, manager, routing, solution)
    else:
        print("No solution found !")
    return solution, manager, routing

```
