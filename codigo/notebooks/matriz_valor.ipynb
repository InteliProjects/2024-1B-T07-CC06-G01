```python
data = {
    'Atributo': ['Custo benefício', 'Qualidade', 'Tecnologia','Personalização','Integração','Suporte ao Cliente'],  # Atributos que servirão como eixo x
    'Atlânticos': [10, 7, 6, 9, 8, 0],  
    'Uber': [7, 7, 8, 7, 8, 7],  
    'Waze': [9, 8, 9, 8, 9, 6], 
    'Fedex': [6, 9, 7, 6, 8, 8], 
    'DHL': [7, 8, 7, 6, 8, 7]
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)

# Transformando o DataFrame para o formato longo (necessário para 'seaborn.lineplot')
df_long = df.melt(id_vars='Atributo', var_name='Empresa', value_name='Nota')

# Configure o estilo do gráfico
sns.set(style="darkgrid")
plt.figure(figsize=(10, 6))  # Definindo o tamanho do gráfico

# Criando o gráfico de linhas, diferenciando por empresa
sns.lineplot(data=df_long, x='Atributo', y='Nota', hue='Empresa', marker='o')

# Adicionando título e rótulos aos eixos
plt.title('Notas para Diferentes Atributos por Empresa', fontsize=16)
plt.xlabel('Atributo', fontsize=12)
plt.ylabel('Nota', fontsize=12)
plt.legend(title='Empresa')
plt.show()  # Exibindo o gráfico
```
